```python
import pandas as pd
import matplotlib.pyplot as plt

# ma'lumotlar
malumotlar = {'oylar': ['Yanvar', 'Fevral', 'Mart', 'Aprel', 'May', 'Iyun', 'Iyul', 'Avgust', 'Sentabr', 'Oktabr', 'Noyabr', 'Dekabr'],
               'Navoiy haroratlar': [2.0, 4.4, 11.7, 18.8, 24.5, 29.7, 32.0, 30.8, 25.3, 17.5, 9.9, 3.9]}

# ma'lumotlar tugmasi
df = pd.DataFrame(malumotlar)

# diagrammani yaratish
fig, ax = plt.subplots()

# haroratlar ustiga diagrammani yaratish
ax.plot(df['oylar'], df['Navoiy haroratlar'], label='Navoiy')

# aksessuarlar
ax.set_xlabel('Oylar')
ax.set_ylabel('O\'rtacha harorat, °C')
ax.set_title(' Navoiy viloyatini yillik o\'rtacha harorati')
ax.legend()

# histogrammani ko'rsatish
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ma'lumotlar
weblar = ['jovian.com', 'google.com', 'facebook.com', 'olx.uz', 'avito.uz']
oylar = ['Yanvar', 'Fevral', 'Mart', 'Aprel', 'May', 'Iyun', 'Iyul', 'Avgust', 'Sentabr', 'Oktabr', 'Noyabr', 'Dekabr']
foydalanuvchilar = np.array([[1500, 2000, 2300, 2500, 2700, 3000, 3100, 3000, 2800, 2700, 2200, 2000], 
                             [3500, 4200, 4900, 5200, 5600, 6000, 6100, 5900, 5700, 5500, 5000, 4800],
                             [5000, 5600, 5900, 6200, 6400, 6700, 7000, 6800, 6500, 6300, 6000, 5800],
                             [1800, 2000, 2100, 2200, 2500, 2800, 2900, 2800, 2600, 2500, 2000, 1800],
                             [1700, 1900, 2000, 2100, 2400, 2700, 2800, 2600, 2400, 2300, 2000, 1800]])

# diagrammani yaratish
fig, ax = plt.subplots()

# barchart yaratish
width = 0.15
for i, weblar in enumerate(weblar):
    ax.bar(np.arange(len(foydalanuvchilar[0])) + (i-2) * width, foydalanuvchilar[i], width, label=weblar)

# aksessuarlar
ax.set_xticks(np.arange(len(oylar)))
ax.set_xticklabels(oylar, rotation=45)
ax.set_xlabel('Oylar')
ax.set_ylabel('Foydalanuvchilar soni')
ax.set_title('O\'zbekistonda bir nechta saytlar uchun yillik foydalanuvchilar soni')
ax.legend()

# histogrammani ko'rsatish
plt.show()
```

```python
import matplotlib.pyplot as plt

# ma'lumotlar
kasalliklar = ['Kanser', 'Diabet', 'Koronavirus', 'Osteoparoz', 'Alzheimer']
olimlar = [1200, 800, 3000, 400, 500]

# piechart yaratish
fig, ax = plt.subplots()
ax.pie(olimlar, labels=kasalliklar, autopct='%1.1f%%')

# aksessuarlar
ax.set_title('O\'zbekistonda bir nechta kasallik tufayli o\'tkazilgan yillik o\'limlar soni')

# piechartni ko'rsatish
plt.show()
```
